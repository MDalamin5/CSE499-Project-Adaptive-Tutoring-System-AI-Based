```python
import json
import pandas as pd

# Function to convert dataset into Phi-4 format
def convert_to_phi4_format(conversations):
    formatted_conversations = []

    for conv in conversations:
        conversation = conv["conversation"]
        formatted_convo = ""

        # System message to establish the assistant's role
        formatted_convo += "<|im_start|>system<|im_sep|>\nYou are a helpful AI math solver assistant. Solve this problem step by step.\n<|im_end|>\n"

        for turn in conversation:
            role = turn["role"]
            message = turn["message"]

            # Fix escape sequences
            message = message.replace("\\(", "(").replace("\\)", ")")

            if role == "student":  # Convert student role to 'user'
                formatted_convo += "<|im_start|>user<|im_sep|>\n" + message + "\n<|im_end|>\n"
            elif role == "system":  # Convert system role to 'assistant'
                formatted_convo += "<|im_start|>assistant<|im_sep|>\n" + message + "\n<|im_end|>\n"

        formatted_conversations.append({"text": formatted_convo})  # Store in a dictionary format

    return formatted_conversations

# Load your .jsonl dataset using Pandas
input_file = "data-3.jsonl"  # Change this to your actual filename
df = pd.read_json(input_file, lines=True)

# Convert dataset to Phi-4 format
formatted_data = convert_to_phi4_format(df.to_dict(orient="records"))

# Save the formatted dataset to a new .jsonl file
output_file = "phi4_formatted_data.jsonl"
with open(output_file, "w", encoding="utf-8") as f:
    for entry in formatted_data:
        f.write(json.dumps(entry, ensure_ascii=False) + "\n")

print(f"Dataset successfully converted and saved to '{output_file}'.")
```
